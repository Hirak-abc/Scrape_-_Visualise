```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define file path
file_path = r"E:\\DHP\\Sub_Fold\\Final_Pro"  # Ensure the file exists

# Check if the file exists
if not os.path.exists(file_path):
    print(f"Error: File not found at {file_path}")
    exit()

# Load the Excel file
try:
    df = pd.read_excel(file_path, engine="openpyxl")  # Use 'openpyxl' for .xlsx files
except Exception as e:
    print(f"Error reading the file: {e}")
    exit()

# Ensure the correct column names
expected_columns = ["Date", "Tags"]  # Case-sensitive
for col in expected_columns:
    if col not in df.columns:
        print(f"Error: Column '{col}' not found in the file. Available columns: {df.columns.tolist()}")
        exit()

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Drop rows where date conversion failed
df = df.dropna(subset=["Date"])

# Extract year from Date
df["Year"] = df["Date"].dt.year

# Ask for user input
tag1 = input("Enter first tag: ").strip().lower()
tag2 = input("Enter second tag: ").strip().lower()

# Count occurrences of each tag per year
tag1_counts = df[df["Tags"].str.contains(tag1, case=False, na=False)].groupby("Year").size()
tag2_counts = df[df["Tags"].str.contains(tag2, case=False, na=False)].groupby("Year").size()

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(tag1_counts.index, tag1_counts.values, marker="o", linestyle="-", label=tag1, color="blue")
plt.plot(tag2_counts.index, tag2_counts.values, marker="o", linestyle="-", label=tag2, color="red")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Number of Questions")
plt.title("Stack Overflow Tags Trend Over Past Three Years")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys  # Import sys for better exit handling

# Define the correct file path (ensure it includes the full filename and extension)
file_path = r"E:\DHP\Sub_Fold\Final_Pro.xlsx"  # Ensure the correct extension

# Check if the file exists
if not os.path.isfile(file_path):
    print(f"Error: File not found at {file_path}")
    sys.exit()

# Load the Excel file
try:
    df = pd.read_excel(file_path, engine="openpyxl")  # Use 'openpyxl' for .xlsx files
except Exception as e:
    print(f"Error reading the file: {e}")
    sys.exit()

# Ensure the correct column names
expected_columns = ["Date", "Tags"]  # Case-sensitive
missing_columns = [col for col in expected_columns if col not in df.columns]

if missing_columns:
    print(f"Error: Missing columns {missing_columns}. Available columns: {df.columns.tolist()}")
    sys.exit()

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Drop rows where date conversion failed
df = df.dropna(subset=["Date"])

# Extract year from Date
df["Year"] = df["Date"].dt.year

# **Filter data to only include years from 2022 to 2024**
df = df[(df["Year"] >= 2022) & (df["Year"] <= 2024)]

# Ask for user input
tag1 = input("Enter first tag: ").strip().lower()
tag2 = input("Enter second tag: ").strip().lower()

# Count occurrences of each tag per year
tag1_counts = df[df["Tags"].str.contains(tag1, case=False, na=False)].groupby("Year").size()
tag2_counts = df[df["Tags"].str.contains(tag2, case=False, na=False)].groupby("Year").size()

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(tag1_counts.index, tag1_counts.values, marker="o", linestyle="-", label=tag1, color="blue")
plt.plot(tag2_counts.index, tag2_counts.values, marker="o", linestyle="-", label=tag2, color="red")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Number of Questions")
plt.title("Stack Overflow Tags Trend (2022-2024)")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
